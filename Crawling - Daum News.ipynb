```python
# 수집 대상: https://news.daum.net/
# 페이지로 요청 -> 응답 

from urllib.request import urlopen

#http 요청
req = urlopen("https://news.daum.net/")
print(req.getcode()) #200= 성공코드

html = req.read() #응답 읽어오기

#인코딩이 깨지므로 ->  UTF-8 로 전환 
html=html.decode("UTF-8")
print(html)

```

```python
# 객체화 
from bs4 import BeautifulSoup
#Tree 객체화 
soup = BeautifulSoup(html,"html.parser") #  Soup 생성
# print (soup.prettify)

#문서 정보의 확인
print("문서 제목:", soup.title, type(soup.title))
print ("태그명:", soup.title.name) #태그 이름
print(soup.title.text, soup.title.string)
```
