```python
# mtcars dataset을 이용한 산점도 그래프 (scatter plot)
import pandas as pd
mtcars=pd.read_csv('./data/mtcars.csv')
mtcars
```

```python
from IPython.display import display
#EDA
print("Size:",mtcars.shape) 
#레코드는 32개 컬럼 수는 12개
#위쪽 데이터 확인
display(mtcars.head())
#아래쪽 데이터 확인
display(mtcars.tail())
```

```python
#기초 통계량 확인
mtcars.describe()
#mpg=mile per gallon, wt=weight
```

```python
#mpg((연비)와 wt(중량) -> 산점도 그래프를 그리면 상관관계를 알수 있음 
plt.scatter(mtcars['wt'],mtcars['mpg']) #(x축, y축)

plt.xlabel('Weight')
plt.ylabel('Miles per Gallon')
plt.title('Weight vs. MPG')

plt.show()
```

```python
#상관계수 (Correlation Coefficient)
print("상관계수 (wt vs mpg)")
mtcars.corr() #모든 수치 변수들 사이의 상관 계수 
#wt변수와  mpg변수 사이의 상관계수 -> 역상관 관계임을 확인
mtcars['wt'].corr(mtcars['mpg']) 
```

```python
#상자그림 (Box Plot)
plt.boxplot(mtcars['mpg']) #단일변수의 box plot
# plt.boxplot([mtcars['mpg'],mtcars['wt']])
plt.show()

print("mpg 기초통계량")
display(mtcars['mpg'].describe())
```

```python
#분위수 확인
quantiles=mtcars.quantile([0.25,0.5,0.75]) #모든 변수의 분위수
display(quantiles)
```

```python
#mpg의 1,3사분위수
first_quantile=mtcars['mpg'].quantile(0.25) #1사분위수
third_quantile=mtcars['mpg'].quantile(0.75) #3사분위수
print("1사분위수: {}, 3사분위수: {}".format(first_quantile,third_quantile))

#IQR (Interquantile Range): Q3-Q1
iqr=third_quantile-first_quantile
print("IQR of mpg:", iqr)
```

```python
#mpg의 상단 극단치 경계
#Q3+iqr*1.5

border_top=third_quantile + irq *1.5
print("상단 극단치 경계:", border_top)

#mpg의 하단 극단치 경계
#Q1-iqr*1.5
border_bottom= first_quantile -irq *1.5
print("하단 극단치 경계:", border_bottom)
```

```python
#mpg가 극단치를 가진 자동차? 
#mpg가 상단 극단치를 초과한 레코드 
# mtcars['mpg'] >border_top #boolean 연산
outliers=mtcars[mtcars['mpg']>border_top]
print("Outliers")
display(outliers)
```
